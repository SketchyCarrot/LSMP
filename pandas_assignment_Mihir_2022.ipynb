<h3>2. Given the technologies of all organizations try to give them a score of how many tech stacks you are familiar with.</h3>

```python
# Taking all entries from Technology column and creating a new DataFrame by splitting.

techs = pd.DataFrame(df.Technologies.str.split(" - "))     # Removing the dash
techs
```

```python
# Combining all the lists in the column

technologies = []
for i in range(len(techs)):
        technologies += techs.loc[i, "Technologies"]     # Using loc function to access individual entries
technologies
```

```python
# Creating a DataFrame from the previous combined list of all entries.

technologies_involved = pd.DataFrame({"Individual": technologies})
technologies_involved
```

```python
# Finding out all the unique technologies used by all the Organizations.
# Total number of unique technoligies is 268.

technologies_involved.Individual.unique()
```

```python
len(technologies_involved.Individual.unique())
```

```python
# Making a list of all technologies I currently know after viewing the unique technologies list.

known_technologies = ["python", "html", "c++", "c", "css", "matlab", "data analysis", "numpy", "web development", "git", "jupyter", "Python", "Numpy", "HTML", "CSS", "Matlab", "Data analysis", "Jupyter", "c/c++"]
points = []     # An empty string to assign points

# Using for loop to increase points for every technology known.

for i in range(len(techs)):
        points.append(0)     # Adding 0 point for each entry. Total 195 entries will be made, equal to the number of Organizations
        for t in known_technologies:
                for j in range(len(techs.loc[i, "Technologies"])):     # Going through all technologies of an Organization
                        if t == techs.loc[i, "Technologies"][j]:
                                points[i] += 1
                        else:
                                points[i] += 0
points




# techs["points"] = 0
# for i in range(len(techs)-1):
#         for t in range(len(techs["Technologies"])):
#                 for i in range(len(techs["Technologies"][t])-1):
#                         for j in range(len(known_technologies)):
#                                 if known_technologies[j] == techs["Technologies"][i]:
#                                         techs["points"][t] += 1

```

```python
# Points can added to the DataFrame.

len(points)
```

```python
# Making a new column in the original DataFrame for points.

df["Points"] = points
df
```

<h3>3. Sort the table as per the assigned score.</h3>

```python
# Using the sort_values function.

sorted_df = df.sort_values(by="Points", ascending=False)
sorted_df
```
